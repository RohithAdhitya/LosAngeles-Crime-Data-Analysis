# **3. Data Cleaning:**

```python
#3.1 Identifying and handling the missing data appropriately.

mv = df.isnull().sum()
print("Missing Data:\n", mv)
```

```python
# Filling '0' for int and float values

df['Mocodes'] = df['Mocodes'].fillna(0)
df['Crm Cd 1'] = df['Crm Cd 1'].fillna(0)
df['Crm Cd 2'] = df['Crm Cd 2'].fillna(0)
df['Crm Cd 3'] = df['Crm Cd 3'].fillna(0)
df['Crm Cd 4'] = df['Crm Cd 4'].fillna(0)
df['Premis Cd'] = df['Premis Cd'].fillna(0)
df['Weapon Used Cd'] = df['Weapon Used Cd'].fillna(0)
```

```python
# Filling 'Not Available' for Object

df['Vict Sex'] = df['Vict Sex'].fillna('Not Available')
df['Vict Descent'] = df['Vict Descent'].fillna('Not Available')
df['Premis Desc'] = df['Premis Desc'].fillna('Not Available')
df['Weapon Desc'] = df['Weapon Desc'].fillna('Not Available')
df['Cross Street'] = df['Cross Street'].fillna('Not Available')
```

```python
#3.2 Checking and removing duplicate rows.

dup_rows = df[df.duplicated()]
print(f"Duplicate Rows except first occurrence: {dup_rows}")
```

```python
#3.3 Converting data types if needed (e.g., dates to date format, numerical values to appropriate numeric types).

df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])
df['TIME OCC'] = pd.to_datetime(df['TIME OCC'], format='%H%M', errors='coerce')
df['Premis Cd'] = df['Premis Cd'].astype(int)
df['Weapon Used Cd'] = df['Weapon Used Cd'].astype(int)
```

```python
#3.4 Dealing with outliers

numerical_cols = df.select_dtypes(include=[np.number]).columns

plt.figure(figsize=(15, 6))
sns.boxplot(data=df[numerical_cols])
plt.title("Box Plot - Before Handling Outliers")
plt.tight_layout()
plt.show()
```

```python
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = (df[col] < lower_bound) | (df[col] > upper_bound)
    df[col] = np.where(outliers, df[col].median(), df[col])
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(data=df[numerical_cols])
plt.title("Box Plot - After Handling Outliers")
plt.tight_layout()
plt.show()
```

```python
#4.1 Visualizing overall crime trends from 2020 to the present year.

monthly_crime_counts = df.groupby(df['DATE OCC'].dt.to_period('M')).size()

plt.figure(figsize=(12, 6))
monthly_crime_counts.plot(kind='line', marker='o', linestyle='-')
plt.title('Monthly Crime Trends from 2020 to Present Year')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#4.2 Analyzing and visualizing seasonal patterns in crime data.

df['Year'] = df['DATE OCC'].dt.year
df['Month'] = df['DATE OCC'].dt.month

monthly_crime_counts = df.groupby(['Year', 'Month']).size()
monthly_crime_counts = df.groupby(['Year', 'Month']).size().reset_index(name='Count')

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_crime_counts, x='Month', y='Count', hue='Year', marker='o', palette='Set1')
plt.title('Seasonal Patterns in Crime Data')
plt.xlabel('Month')
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(1, 13), month_labels)
plt.ylabel('Number of Crimes')
plt.legend(title='Year',bbox_to_anchor=(1, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#4.3 Identifying the most common type of crime and its trends over time.

crime_counts = df['Crm Cd Desc'].value_counts()

most_common_crime = crime_counts.idxmax()
most_common_crime_count = crime_counts.max()
most_common_crime_data = df[df['Crm Cd Desc'] == most_common_crime]
most_common_crime_counts = most_common_crime_data.groupby(['Year', 'Month']).size()
plt.figure(figsize=(12, 6))
most_common_crime_counts.plot(kind='line', marker='o', label=most_common_crime,color='red')

plt.title(f'Trends of the Most Common Crime: {most_common_crime}')
plt.xlabel('Year & Month')
x_ticks = [f"{year}-{month:02}" for year, month in most_common_crime_counts.index]
plt.xticks(range(len(x_ticks)), x_ticks, rotation=75)
plt.ylabel('Number of Crimes')
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#4.6 Analyzing the relationship between the day of the week and the frequency of certain types of crimes.

# Function to categorize crimes
def categorize_crime(crime_description):
    if 'theft' in crime_description.lower():
        return 'Property Crimes'
    elif 'assault' in crime_description.lower():
        return 'Violent Crimes'
    elif 'burglary' in crime_description.lower():
        return 'Property Crimes'
    elif 'robbery' in crime_description.lower():
        return 'Violent Crimes'
    else:
        return 'Other Crimes'

df['Crime Category'] = df['Crm Cd Desc'].apply(categorize_crime)
df['Day of Week'] = df['DATE OCC'].dt.day_name()
crime_counts = df.groupby(['Crime Category', 'Day of Week']).size().unstack(fill_value=0)

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='Day of Week', hue='Crime Category', palette='magma')
plt.title('Crime Categories by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Crime Category',bbox_to_anchor=(1, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#10.Predicting Future Trends:
#Employ time series forecasting methods, such as ARIMA or Prophet, to predict future crime trends based on historical data.
#Consider incorporating relevant external factors into your models.

from statsmodels.tsa.arima.model import ARIMA

ts_d = df.groupby('DATE OCC').size().reset_index()
ts_d.columns = ['ds', 'y']
ts_d.set_index('ds', inplace=True)
order = (1, 1, 1)
arima_model = ARIMA(ts_d, order=order)
arima_results = arima_model.fit()
forecast_steps = 365
forecast = arima_results.get_forecast(steps=forecast_steps)

plt.figure(figsize=(12, 6))
plt.plot(ts_d.index, ts_d['y'], label='Observed')
plt.plot(forecast.predicted_mean.index, forecast.predicted_mean, color='red', label='ARIMA Forecast')
plt.title('Crime Forecast using ARIMA')
plt.xlabel('Date')
plt.ylabel('Crime Count')
plt.legend()
plt.show()

```
